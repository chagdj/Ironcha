```python
import pandas as pd
import requests

url = 'https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_1983'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())

```

```python
print(response.status_code)
```

```python
tables = soup.find_all('table')  # List all the tables

print(f"Number of tables: {len(tables)}")  # Number of tables
for i, table in enumerate(tables):
    print(f"Table {i}:")  # Display the number of tables
    print(table.prettify())  
```

```python
for song in soup.find_all("td"):
    songs = print(song.get_text())

```

```python
songs = []  

for song in soup.find_all("td"):
    song_text = song.get_text(strip=True)  # Get the text and strip whitespace
    if song_text:  # Check if song_text is not empty
        songs.append(song_text)  # Add the new text to the songs list

print("Titres de chansons valides :")
for title in songs:
    print(title)
```

```python
import re
joined_songs = ''.join(songs)

# Use regex
song_titles = re.findall(r'"(.*?)"\s*([\w\s&]+)', joined_songs)


valid_titles = [f'{title} by {artist}\n' for title, artist in song_titles]


print("Titres de chansons valides :\n")
for title in valid_titles:
    print(title)

```
